```python
resources = {
'sheep': {'n': 4, 'color': 'yellowgreen', 'code': 'SH'},
'wheat': {'n': 4, 'color': 'khaki', 'code': 'WH'},
'wood': {'n': 4, 'color': 'peru', 'code': 'WO'},
'brick': {'n': 3, 'color': 'firebrick', 'code': 'BR'},
'stone': {'n': 3, 'color': 'lightgray', 'code': 'ST'},
'desert': {'n': 1, 'color': 'orange', 'code': 'DE'},
}

rolls = {2:1, 3:2, 4:2, 5:2, 6:2, 8:2, 9:2, 10:2, 11:2, 12:1}
```

```python
class Node(object):
    counter=0
    def __init__(self, r, c):
        self.r = r
        self.c = c
        self.counter = Node.counter
        self.hexes = []
        Node.counter += 1
        
    def add_hex(self, hex_):
        if hex_ not in self.hexes:
            self.hexes.append(hex_)

    def __repr__(self):
        return 'Node %s %s' % (self.counter, ' '.join(_.__repr__() for _ in self.hexes))
    
class Hex(object):
    counter=0
    def __init__(self, r, c):
        self.r = r
        self.c = c
        self.counter = Hex.counter
        self.resource = reso_ar.pop()
        self.color = filter(lambda x: x['code'] == self.resource, resources.values())[0]['color']
        if self.resource != 'DE':
            self.roll = roll_ar.pop()
        else:
            self.roll = ''
                
        Hex.counter += 1
    def __repr__(self):
        return '%s %s' % (self.resource, self.roll)

class Grid(object):
    def __init__(self, nrows, ncols):
        self.grid = np.full((nrows, ncols), -1)
        self.index = 0
        self.objs = []
    
    def add_node(self, r, c):
        self.grid[r, c] = self.index
        self.objs.append(Node(r, c))
        self.index += 1

    def add_hex(self, r, c):
        if r%2:
            self.grid[r, c] = self.index
            self.grid[r+1, c] = self.index
            self.objs.append(Hex(r, c))
            self.index += 1 
            
    def process_objs(self):
        for obj in self.objs:
            if type(obj) == Hex:
                continue
            nbs = get_neighbors(obj.r, obj.c)
            for n in nbs:
                if type(n) == Hex:
                    obj.add_hex(n)
                    
    def get_neighbors(self, r, c):
        out = []
        for dr, dc in product([-1, 0, 1], repeat=2):
            if dr==0 and dc==0:
                continue

            r_ = r + dr
            c_ = c + dc        

            if r_ < 0 or c_ < 0 or r_ >= self.grid.shape[1] or c_ >= self.grid.shape[1]:
                continue
            idx = self.grid[r_, c_]
            if idx >= 0:
                out.append(self.objs[int(idx)])

        return list(set(out))

    def to_table(self):
        out = '<table>'
        for r in self.grid:
            out += '<tr>'
            for c in r:            
                if c < 0:
                    color = 'black'
                    label = ''
                else:
                    obj = self.objs[int(c)]
                    if type(obj) == Node:
                        color = 'slategray'
                        label = obj.__repr__()
                    else:
                        color = obj.color
                        label = obj.__repr__()
                        
                out += '<td style="background-color: %s">%s</td>' % (color, label)
            out += '</tr>'
        out += '</table>'

        return HTML(out)
```

```python
reso_ar, roll_ar = [], []
map(lambda x: reso_ar.extend(x['n']*[x['code']]), resources.values())
map(lambda x: roll_ar.extend(x[1]*[x[0]]), rolls.items())
shuffle(reso_ar)
shuffle(roll_ar)

# Number of nodes in start row
start = 3

# Number of nodes in middle row
middle = 6

alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

nrows = 4*(middle - start)
ncols = 2*middle - 1

grid = Grid(nrows, ncols)

# index to start drawing nodes
start_index = middle - start

# when to start bringing the board back in
half = 2*(middle-start)
index_switch = 2

print 'nrows: %s, ncols: %s, half: %s' % (nrows, ncols, half)

for r in range(nrows):
    
    #always start with a node
    node = 1
    
#     print 'row: %d, start_index: %d, hex_switch: %d' % (r, start_index, hex_switch)
    
    for c in range(ncols):
        if (c >= start_index) and (c < (ncols-start_index)):
            if node:
                # its a node
                grid.add_node(r, c)
            else:
                # its a hex  
                if r != 0 and r != (nrows-1): 
                    grid.add_hex(r, c)
            node = not node
    
    # controls whether to move the offset in or out or stay the same
    if r > half-1:
        if (index_switch%2 == 0):            
            start_index += 1
        index_switch += 1
    else:
        if (index_switch%2 == 0):            
            start_index -= 1
        index_switch += 1

grid.process_objs()
grid.to_table()
```
